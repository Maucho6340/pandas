#### El equipo de ML ha llegado con algunas demandas de última hora que debemos resolver en este momento de análisis exploratorio. Estas demandas son las siguientes:

1.- Calcular el promedio de habitaciones por departamento.

2.- Verificar cuántas colonias únicas existen en nuestra base de datos.

3.- Analizar qué colonias tienen el promedio de alquiler más alto.

4.- Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.

```python
# 1.- Calcular el promedio de habitaciones por departamento.
# Podemos seleccionar la columna de interés y aplicar el método mean():

df['Habitaciones'].mean()
```

```python
# 2.- Verificar cuántas colonias únicas existen en nuestra base de datos.
#  Una forma de verificar cuántos barrios únicos existen en la base de datos
# sería utilizando el método unique(), que nos devolverá una lista con los barrios únicos.
# Luego, podemos usar la función len para contar cuántos elementos hay en esa lista:

len(df['Colonia'].unique())
```

```python
df['Colonia'].nunique()
```

```python
df['Colonia'].unique()
```

```python
# 3.- Analizar qué colonias tienen el promedio de alquiler más alto.
# Agrupar nuestra base de datos por la columna "Colonia".
# Aplicar el método mean.
# Colocar el resultado en un DataFrame y ordenarlo.

df.groupby('Colonia')[['Valor']].mean().sort_values('Valor', ascending=False)
```

```python
# 4.- Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.
# Como queremos crear un gráfico con solo las 5 colonias que tienen el promedio de alquiler más alto, podemos usar
# .head() para seleccionar las 5 primeras filas del DataFrame que creamos en el desafío 3 y guardar eso en una variable:
df_colonias = df.groupby('Colonia')[['Valor']].mean().sort_values('Valor', ascending=False).head()
df_colonias.plot(kind='barh', figsize=(8,4), color='grey');
```

## **Remover registros inconsistentes**

Estos registros son inconsistentes, por lo tanto, debemos eliminarlos de nuestra base de datos.

## APLICAR FILTROS
En este momento, el equipo de ML desea evaluar solo propiedades en escenarios específicos, como:

1.-Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

2.- Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

Con base en esto, necesitamos aplicar los filtros necesarios para obtener esta información específica.

## 1.-Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

```python
seleccion_1 = df['Habitaciones']==1
```

```python
seleccion_1
```

```python
df[seleccion_1]
```

```python
seleccion_2 = df['Valor']<4200
```

```python
df[seleccion_2]
```

```python
df[seleccion_2].max()
```

```python
filtro_1=(seleccion_1)&(seleccion_2)
df[filtro_1]
```

##2.- Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
filtro_2=(df['Habitaciones']>=2)&(df['Valor']<10500)&(df['Area']>70)
```

```python
df[filtro_2]
```

```python
df_2=df[filtro_2]
```

```python
df_2
```
