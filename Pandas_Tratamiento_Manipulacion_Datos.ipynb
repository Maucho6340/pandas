##### **Proyecto Desafío 1: Ventas Online**

El objetivo de este proyecto es analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que más gastaron en productos durante los 5 días de ventas, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esta semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.

La base de datos utilizada en este análisis es dados_vendas_clientes.json y contiene información importante sobre los clientes, como el nombre registrado del cliente, el monto total pagado al momento de la compra y el día de la compra.

Sabiendo esta información, el desafío del proyecto 1: ventas online será abrir la base de datos con Pandas y aplicar json_normalize.

```python

import pandas as pd
import requests
from io import StringIO # Necesitamos StringIO para leer la cadena de texto como un archivo

url = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json"

#data_clientes = pd.read_json(url)
#data_cleintes.head()

try:
    # Realiza una solicitud GET a la URL.
    # Es buena práctica incluir un User-Agent para simular un navegador,
    # lo que a veces ayuda a evitar bloqueos 403.
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    response = requests.get(url, headers=headers)

    # Lanza una excepción si la solicitud HTTP no fue exitosa (por ejemplo, 403, 404, 500)
    response.raise_for_status()

    # Obtiene el contenido del JSON como una cadena de texto
    json_data = response.text

    # Usa StringIO para tratar la cadena de texto como un archivo
    # y luego pd.read_json puede leerlo.
    data_clientes = pd.read_json(StringIO(json_data))

    # Muestra las primeras filas del DataFrame
    data_clientes.head()

except requests.exceptions.HTTPError as e:
    print(f"Error HTTP: {e}")
    print("Asegúrate de que la URL es correcta y que tienes permiso para acceder al recurso.")
except requests.exceptions.ConnectionError as e:
    print(f"Error de conexión: {e}")
    print("Verifica tu conexión a internet o la disponibilidad del servidor.")
except requests.exceptions.Timeout as e:
    print(f"Tiempo de espera agotado: {e}")
    print("La solicitud tardó demasiado en responder.")
except requests.exceptions.RequestException as e:
    print(f"Error inesperado al hacer la solicitud: {e}")
except ValueError as e:
    print(f"Error al parsear el JSON: {e}")
    print("El contenido descargado podría no ser un JSON válido.")
```

```python
# normalizar archivo json
data_clientes = pd.json_normalize(data_clientes['dados_vendas_clientes'])
data_clientes.head()
```
