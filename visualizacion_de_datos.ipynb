### 01 Conociendo la biblioteca matplotlib

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv') #lectura y asignación del archivo a un DataFrame "df"
```

```python
df.head(30) # Visualización del DF
```

```python
df.shape # tamaño del DF
```

```python
df.set_index('Country',inplace=True) # Ponemos 'Country' como índice (index)
```

```python
df #revisamos df
```

```python
anos = list(map(str,range(1980,2014))) # creamos lista con años
```

```python
anos
```

```python
colombia = df.loc['Colombia',anos] # creamos lista con los años y la cantidad de inmigrantes colombianos
```

```python
colombia
```

```python
# creamos diccionario con los años y la cantidad de inmigrantes colombianos
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist(),
            }
```

```python
col_dict # revisamos diccionario: años - inmigrantes colombianos
```

```python
datos_col = pd.DataFrame(col_dict) # convertimos el diccionario en un DF
```

```python
datos_col.tail()
```

```python
import matplotlib.pyplot as plt # importamos la biblioteca para graficar
```

```python
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.show() # mostrar el gráfico
```

```python
# mejoramos lectura de datos en escala eje "X" con xticks, en "Y" no es necesario
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014']) # datos del eje "X"
plt.show() # mostrar el gráfico
```

```python
# Agregamos títulos para mejor entendimiento del gráfico
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014']) # datos del eje "X"
#Colocando títulos
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('n° de inmigrantes')
# Mostrar
plt.show() # mostrar el gráfico. NOTA:siempre debe ser el último comando
```

### 02 Creando figuras con matplotlib

```python
# Creamos un objeto (variable) que una  lienzo (figure) con la gráfica (plot)
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.show()
```

```python
# Aparece una gráfica sin escalado en eje "X"
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Creamos una función
# que se aplica a "ax",
# en el eje "x"
# mediante método de localización múltiple set-major-locator
# Y pone ticks eje "x" con valores de 5 en 5
# Es lo mismo, pero automatizado que "plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])"
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
# Agregando títulos
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# Agregamos títulos al objeto antes de la última línea de código plt.show
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('n° de inmigrantes')
plt.show()
```

```python
# Gráfico de Caja
fig, axs = plt.subplots(1,2,figsize=(10,3))
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('n° de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la Inmigración de colombianos hacia Canadá\n 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('n° de inmigrantes')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
# Aplicamos iteración para etiquetas
fig,axs=plt.subplots(2,2,figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.5) #fija valores para espacio horizontal y vertical entre gráficos
axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brazil', anos])
axs[0,1].set_title('Brazil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat: # para etiquetado
 ax.xaxis.set_major_locator(plt.MultipleLocator(5))
 ax.set_xlabel('Año')
 ax.set_ylabel('Número de Inmigrantes')
 ax.grid()
plt.show()
```

```python
# Normalizamos la ecala de "Y"
fig,axs=plt.subplots(2,2,figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.5) #fija valores para espacio horizontal y vertical entre gráficos
axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brazil', anos])
axs[0,1].set_title('Brazil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

# Misma escala para los 4 gráficos
ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
 ax.xaxis.set_major_locator(plt.MultipleLocator(5))
 ax.set_xlabel('Año')
 ax.set_ylabel('Número de Inmigrantes')
 ax.grid()
plt.show()
```

```python
# Agregamos título general a la gráfica
fig,axs=plt.subplots(2,2,figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
fig.suptitle('Inmigración de países Sudamericanos hacia Canadá\n desde1980 a 2013') # Título general

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brazil', anos])
axs[0,1].set_title('Brazil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
 ax.xaxis.set_major_locator(plt.MultipleLocator(5))
 ax.set_xlabel('Año')
 ax.set_ylabel('Número de Inmigrantes')
 ax.grid()
plt.show()
```

### 02 Personalizando gráficos con matplotlib

```python
# Cambiando el tamaño de las fuentes a partir del gráfico de inmigración colombiana
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# Agregamos títulos al objeto antes de la última línea de código plt.show
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('n° de inmigrantes')
plt.show()
```

```python
# Uso de fontsize en títulos
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# cambio tamaño fuente
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013', fontsize=18) # fuente tamaño 18
ax.set_xlabel('Año', fontsize=14) # fuente tamaño 14
ax.set_ylabel('n° de inmigrantes', fontsize=14) # fuente tamaño 14

plt.show()
```

```python
# Uso de fontsize en números
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013', fontsize=18) # fuente tamaño 18
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('n° de inmigrantes', fontsize=14)

# Cambio tamaño números
ax.xaxis.set_tick_params(labelsize=12) # fuente tamaño 12
ax.yaxis.set_tick_params(labelsize=12) # fuente tamaño 12

ax.grid()
plt.show()
```

```python
# Alineación de título
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Cambiar alineación del  título
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013', fontsize=18, loc='left') # Alineado a la izquierda
ax.set_xlabel('Año', fontsize=14)

ax.set_ylabel('n° de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.grid()
plt.show()
```

```python
# Cambiar grosor de la línea (lw) y agregar puntos
fig, ax = plt.subplots(figsize=(8,3))

ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=3, marker='o') #Ancho mas gruesa y puntos circulares

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n 1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('n° de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid()
plt.show()
```

## DESAFÍO 2
Trabajas como Analista de Datos en una empresa minorista y te asignaron la tarea de crear una figura con subgráficos que muestre la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.

¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

**Consejo:**
 Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df = pd.DataFrame(ventas_2022, index=tiendas)
df
```

```python
# Crear la figura y los subgráficos

fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar los espaciados entre los subgráficos

plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Agregar un título general para los subgráficos

fig.suptitle('Ventas en el período de enero a diciembre de 2022 en las tiendas A, B, C y D')

# Agregar los gráficos en cada uno de los subgráficos

axs[0, 0].plot(df.loc['A'])
axs[0, 0].set_title('Ventas en la tienda A')
axs[0, 1].plot(df.loc['B'])
axs[0, 1].set_title('Ventas en la tienda B')
axs[1, 0].plot(df.loc['C'])
axs[1, 0].set_title('Ventas en la tienda C')
axs[1, 1].plot(df.loc['D'])
axs[1, 1].set_title('Ventas en la tienda D')

# Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Número de ventas')

# Mostrar la figura

plt.show()
```
