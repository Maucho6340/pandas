### 01 Conociendo la biblioteca matplotlib

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv') #lectura y asignación del archivo a un DataFrame "df"
```

```python
df.head() # Visualización del DF
```

```python
df.shape # tamaño del DF
```

```python
df.info() # detalle de columnas
```

```python
df.set_index('Country',inplace=True) # Ponemos 'Country' como índice (index)
```

```python
df #revisamos df
```

```python
anos = list(map(str,range(1980,2014))) # creamos lista con años
```

```python
anos
```

```python
colombia = df.loc['Colombia',anos] # creamos lista con los años y la cantidad de inmigrantes colombianos
```

```python
colombia
```

```python
# creamos diccionario con los años y la cantidad de inmigrantes colombianos
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist(),
            }
```

```python
col_dict # revisamos diccionario: años - inmigrantes colombianos
```

```python
datos_col = pd.DataFrame(col_dict) # convertimos el diccionario en un DF
```

```python
datos_col.tail()
```

```python
import matplotlib.pyplot as plt # importamos la biblioteca para graficar
```

```python
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.show() # mostrar el gráfico
```

```python
# mejoramos lectura de datos en escala eje "X" con xticks, en "Y" no es necesario
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014']) # datos del eje "X"
plt.show() # mostrar el gráfico
```

```python
# Agregamos títulos para mejor entendimiento del gráfico
plt.figure(figsize=(10,5)) # tamaño del gráfico
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) # indica primero lo que va en eje "X" y luego en "Y""
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014']) # datos del eje "X"
#Colocando títulos
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('n° de inmigrantes')
# Mostrar
plt.show() # mostrar el gráfico. NOTA:siempre debe ser el último comando
```

### 02 Creando figuras con matplotlib

```python
# Creamos un objeto (variable) que una  lienzo (figure) con la gráfica (plot)
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.show()
```

```python
# Aparece una gráfica sin escalado en eje "X"
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Creamos una función
# que se aplica a "ax",
# en el eje "x"
# mediante método de localización múltiple set-major-locator
# Y pone ticks eje "x" con valores de 5 en 5
# Es lo mismo, pero automatizado que "plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])"
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```
